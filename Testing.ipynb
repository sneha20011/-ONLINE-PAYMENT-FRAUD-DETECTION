```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
user_fraud = data['isFraud']
user_amt = data['amount']

# Group age data by race
user_frame = pd.DataFrame({"fraud":user_fraud,"amount":user_amt})
groups = user_frame.groupby("fraud").groups

# Etract individual groups
fraud = user_amt[groups[0]]
nonfraud = user_amt[groups[1]]

# Perform the ANOVA
stats.f_oneway(fraud,nonfraud)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('amount~fraud', data = user_frame).fit()
                
anova_result = sm.stats.anova_lm(model, typ=2)
print (anova_result)
```

```python
from benfordslaw import benfordslaw

# Initialize
bl = benfordslaw(alpha=0.05)

# Extract election information.
X = data['amount']

# Print
print(X)

# Make fit
results = bl.fit(X)

# Plot
bl.plot(title='First Digit')
```
